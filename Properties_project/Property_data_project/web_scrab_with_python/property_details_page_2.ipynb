```python
import urllib.request
import re
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
api_url='https://api.scrapingdog.com/scrape?api_key=6482e8f843c55505c48bddb2&url=https://www.magicbricks.com/new-projects-Pune/page-2'
```

```python
page=requests.get(api_url)
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
project_name_content=soup.find_all(class_='proHeading')
project_name_content=str(project_name_content)
project_name_content
```

```python
re_project_name=r'>(.*?)<\/strong>'
Project_name=re.findall(re_project_name,project_name_content)
Project_name
```

```python
project_Area_pune_content=soup.find_all(class_='proGroup')
project_Area_pune_content=str(project_Area_pune_content)
project_Area_pune_content
```

```python
re_project_Area_pune=r'>(.*?)<\/a>'
Project_Area_pune=re.findall(re_project_Area_pune,project_Area_pune_content)
Project_Area_pune
```

```python
project_developer_content=soup.find_all(class_='proGroup projDeveloper')
project_developer_content=str(project_developer_content)
project_developer_content
```

```python
re_project_developer=r'by</label> <span>(.*?)<\/span>'
Project_developer=re.findall(re_project_developer,project_developer_content)
Project_developer
```

```python
project_possesion_content=soup.find_all(class_='proDetailsRowElm posByCons')
project_possesion_content=str(project_possesion_content)
project_possesion_content
```

```python
re_project_possesion=r'\b(\w+\s+\w+\s+\d{4})\b'
Project_possesion=re.findall(re_project_possesion,project_possesion_content)
Project_possesion
```

```python
project_2bhk_price_content=soup.find_all(class_='proDescColm3')
project_2bhk_price_content=str(project_2bhk_price_content)
project_2bhk_price_content
```

```python
cleaned_text = re.sub(r'[\n\t]+', '', project_2bhk_price_content)
cleaned_text
```

```python
re_project_2bhk_pricer=r'<\/span>(.*?)<\/div>'

Project_2bhk_price=re.findall(re_project_2bhk_pricer,cleaned_text)
Project_2bhk_price
```

```python
project_carpet_area_content=soup.find_all(class_='proDescColm2')
project_carpet_area_content=str(project_carpet_area_content)
project_carpet_area_content
```

```python
clean=re.sub(r'[\n\t]+','',project_carpet_area_content)
clean
```

```python
re_carpet_area=r'>(.*?)<\/div>'
project_carpet_area=re.findall(re_carpet_area,clean)
project_carpet_area
```

```python
re_carpet_area=r'\d+(?:\s*-\s*\d+)?\s*sqft'
carpet_area=re.findall(re_carpet_area,str(project_carpet_area))
carpet_area
```
